```python
rate=[]
wdata["Release Clause"]=wdata["Release Clause"].fillna("0")

for i in list(wdata["Release Clause"]):
    if("." in i):
        b=i.split(".")[1].replace("€","").replace("M","00000").replace(".","").replace("K","00")
        i=i.split(".")[0].replace("€","")+b
        rate.append(i)
    else:
        i=i.replace("€","").replace("M","000000").replace("K","000")
        rate.append(i)
wdata["Release_Clause"]=rate
wdata.head(15)
```

```python
rate=[]
wdata["Wage"]=wdata["Wage"].fillna("0")

for i in list(wdata["Wage"]):
        i=i.replace("€","").replace("K","000")
        rate.append(i)
wdata["Wage_inEuro"]=rate
wdata.head()
```

```python
wdata.info()
```

```python
wdata['Release_Clause']=pd.to_numeric(wdata['Release_Clause'], errors="coerce")
wdata['Wage_inEuro']=pd.to_numeric(wdata['Wage_inEuro'], errors="coerce")
```
