```python
wdata = data[["Name","Age","Nationality","Overall","Wage","Club","Release Clause"]]
wdata.head()
```

```python
wdata.Wage.unique()
```

```python
len(wdata.Wage.unique())
```

```python
#Comment if wage has direct link with overall average
pd.crosstab(wdata['Wage'], wdata['Overall'])
```

```python
rate=[]
wdata["Release Clause"]=wdata["Release Clause"].fillna("0")

for i in list(wdata["Release Clause"]):
    if("." in i):
        b=i.split(".")[1].replace("€","").replace("M","00000").replace(".","").replace("K","00")
        i=i.split(".")[0].replace("€","")+b
        rate.append(i)
    else:
        i=i.replace("€","").replace("M","000000").replace("K","000")
        rate.append(i)
wdata["Release_Clause"]=rate
wdata.head(15)
```

```python
rate=[]
wdata["Wage"]=wdata["Wage"].fillna("0")

for i in list(wdata["Wage"]):
        i=i.replace("€","").replace("K","000")
        rate.append(i)
wdata["Wage_inEuro"]=rate
wdata.head()
```

```python
wdata.info()
```

```python
wdata['Release_Clause']=pd.to_numeric(wdata['Release_Clause'], errors="coerce")
wdata['Wage_inEuro']=pd.to_numeric(wdata['Wage_inEuro'], errors="coerce")
```

```python
wdata.info()
```

```python
#players and their release clause
adata=wdata.head(55)
plt.figure(figsize=(15,5))
sns.barplot(x=adata['Name'], y=adata['Release_Clause'])
plt.xticks(rotation= 90)
plt.xlabel('Players')
plt.ylabel('co-efficient')
plt.title('Players and release clause')
plt.show()
```

```python
#inference of wageS and overall performance
adata=wdata
plt.figure(figsize=(15,5))
sns.barplot(x=adata['Overall'], y=adata['Wage_inEuro'])
plt.xlabel('Overall Performance')
plt.ylabel('Wages')
plt.title('Overall performance and wages')
plt.show()
```
